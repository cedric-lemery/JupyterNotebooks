# Obtenir la vitesse & l'accélération sur une série de mesure en python

## Données initiales
Supposons que nous ayons une série de mesure de x mesuré à des instants t que nous allons simuler ici de la façon suivante :

```python
import matplotlib.pyplot as plt
t, x=[], []
n=25
for i in range(n):
    ti=i*0.1
    xi=ti**2
    t.append(ti)
    x.append(xi)
plt.title('Evolution de la position au cours du temps')
plt.plot(t, x, 'b')
plt.show()
```

## Obtention de l'évolution de la vitesse
Nous pouvons calculer les coordonnées de la vitesse selon x de la façon suivante :

```python
vx = []
vx.append((x[1]-x[0])/(t[1]-t[0]))
for i in range(1,n-1): #on part de 1 et on va jusqu'à l'avant dernière position
    vxi=(x[i+1]-x[i-1])/(t[i+1]-t[i-1])
    vx.append(vxi)
vx.append((x[n-1]-x[n-2])/(t[n-1]-t[n-2]))
```

Pour en faire le tracé, comme la vitesse n'a pas été calculée à la première et à la dernière mesure, il faut bien penser à enlever la première et la dernière valeur du temps (ce qu'on fait avec *t[1:-1]*):

```python
import matplotlib.pyplot as plt
plt.title('Evolution de la vitesse au cours du temps')
plt.plot(t, vx, 'r')
plt.show()
```

Nous voyons que la vitesse est une fonction croissante du temps. C'est un mouvement rectiligne uniformément accéléré.

*Remarque : on notera qu'il y a un petit effet de bord en 0 et en n du fait des approximations de calculs pour les bords.*

## Obtention de l'évolution de l'accélération
Pour obtenir l'accélération, il faut appliquer la même manipulation sur la vitesse :

```python
ax = []
ax.append((vx[1]-vx[0])/(t[1]-t[0]))
for i in range(1,n-1):
    axi=(vx[i+1]-vx[i-1])/(t[i+1]-t[i-1])
    ax.append(axi)
ax.append((vx[n-1]-vx[n-2])/(t[n-1]-t[n-2]))
    
plt.title("Evolution de l'accélération en fonction du temps")
plt.axis([0,2.5,0,3])
plt.plot(t, ax, 'r')
plt.show()
```

L'accélération est constante, elle est donc représentée par une droite horizontale par rapport au temps.
