# Comment faire une régression linéaire d'une série de mesure ?
Imaginons que nous ayons une série de mesure expérimentale g qui est quasi linéaire en fonction d'un paramètre x :

```python
import numpy as np
import matplotlib.pyplot as plt

nb_point=16
xmax=10
x=np.arange(0,xmax,xmax/nb_point)
g_true=5.*x+1.
g_noise=np.random.randn(len(x))*2
g_mesure=g_true+g_noise
plt.plot(x,g_mesure, '+b', label = "grandeur g_mesure mesurée en fonction de x")
plt.legend()
plt.show()
```

Pour déterminer les valeurs a & b qui permettent d'extrapoler la grandeur mesurée sous la forme g=a.x+b

on procède de la sorte :

```python
fit = np.polyfit(x, g_mesure, 1)
```

fit est un tableau qui contient les valeurs de a & de b :

```python
print(fit)
```

Nous pouvons donc les récupérer pour tracer g_model=a.x+b et le comparer à la valeur mesurée

```python
a=fit[0]
b=fit[1]
plt.plot(x,g_mesure, '+b', label = "grandeur g_mesure mesurée en fonction de x")
plt.plot(x,a*x+b, '-r', label = "g = "+str(a)+".x+"+str(b))
plt.legend()
plt.show()
```
