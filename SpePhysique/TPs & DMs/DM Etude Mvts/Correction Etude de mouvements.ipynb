# Etude de mouvements
Première étape : on importe les bibliothèques nécessaires

```python
import matplotlib.pyplot as plt
import csv
```

## Mouvement 1
On commence par importer les données puis on trace x & y en fonction du temps pour comprendre à quoi on a affaire

```python
source = open('Mvt1.csv', 'r') 
t, x, y = [], [], [] 
for row in csv.reader(source,delimiter=','):
    ti, xi, yi = map(float,row)
    t.append(ti) 
    x.append(xi) 
    y.append(yi) 

plt.title('x & y en fonction du temps pour le mvt 1')
plt.plot(t, x, 'r', label ="x = f(t)")
plt.plot(t, y, 'b', label ="y = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("position", fontsize=12)
plt.legend()
plt.show()
```

Nous voyons que y reste nul tandis que x varie. C'est donc un mouvement rectiligne. Ce que nous pouvons confirmer en traçant le mouvement dans le plan (Oxy) :

```python
plt.title('Trajectoire pour le mvt 1')
plt.plot(x, y, 'r', label ="y en fonction de x")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()
```

Sur la courbe x=f(t), nous voyons que x n'est pas une fonction affine du temps. Ce n'est donc pas un mouvement rectiligne uniforme.

Calculons maintenant l'évolution de la vitesse par rapport au temps, puis traçons vx en fonction du temps :

```python
n=len(x)
vx = []
vx.append((x[1]-x[0])/(t[1]-t[0]))
for i in range(1,n-1): 
    vxi=(x[i+1]-x[i-1])/(t[i+1]-t[i-1])
    vx.append(vxi)
vx.append((x[n-1]-x[n-2])/(t[n-1]-t[n-2]))

plt.title('vx en fonction du temps pour le mvt 1')
plt.plot(t, vx, 'r', label ="vx = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("vitesse", fontsize=12)
plt.legend()
plt.show()
```

Nous observons que vx est une fonction linéaire du temps. Cela suggère que le mouvement est uniformément accéléré. Ce qui est confirmé par le calcul de ax en fonction du temps :

```python
ax = []
ax.append((vx[1]-vx[0])/(t[1]-t[0]))
for i in range(1,n-1):
    axi=(vx[i+1]-vx[i-1])/(t[i+1]-t[i-1])
    ax.append(axi)
ax.append((vx[n-1]-vx[n-2])/(t[n-1]-t[n-2]))

plt.title('ax en fonction du temps pour le mvt 1')
plt.plot(t, ax, 'r', label ="ax = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("accélération", fontsize=12)
plt.legend()
plt.show()
```

Nous observons que l'accélération est constante (on ne prend pas en compte ce qui se passe en i=0 & i=n-1) de valeur 10 m/s².

**Le mouvement 1 est un mouvement rectiligne uniformément accéléré**

## Etude du mouvement 2
Cette fois-ci encore, on commence par importer les données et tracer x & y en fonction du temps :

```python
source = open('Mvt2.csv', 'r') 
t, x, y = [], [], [] 
for row in csv.reader(source,delimiter=','):
    ti, xi, yi = map(float,row)
    t.append(ti) 
    x.append(xi) 
    y.append(yi) 

plt.title('x & y en fonction du temps pour le mvt 2')
plt.plot(t, x, 'r', label ="x = f(t)")
plt.plot(t, y, 'b', label ="y = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("position", fontsize=12)
plt.legend()
plt.show()
```

Nous voyons qu'encore une fois le mouvement est rectiligne :

```python
plt.title('Trajectoire pour le mvt 2')
plt.plot(x, y, 'r', label ="y en fonction de x")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()
plt.show()
```

Mais il est clairement complexe lorsqu'on observe la variation de x par rapport au temps.

Traçons maintenant l'évolution de la vitesse par rapport au temps :

```python
n=len(x)
vx = []
vx.append((x[1]-x[0])/(t[1]-t[0]))
for i in range(1,n-1): 
    vxi=(x[i+1]-x[i-1])/(t[i+1]-t[i-1])
    vx.append(vxi)
vx.append((x[n-1]-x[n-2])/(t[n-1]-t[n-2]))

plt.title('vx en fonction du temps pour le mvt 2')
plt.plot(t, vx, 'r', label ="vx = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("vitesse", fontsize=12)
plt.legend()
plt.show()
```

Nous observons 3 phase pour ce mouvement :
1. initialement le **mouvement est rectiligne uniforme** de vitesse 10 m/s;
2. ensuite le **mouvement est uniformément accéléré** car la vitesse est une fonction affine de pente croissante par rapport au temps;
3. puis, le **mouvement est uniformément ralenti** car la vitesse est une fonction positive, affine de pente décroissante par rapport au temps. La vitesse est dans le sens Ox mais l'accélération est en sens inverse;
4. enfin, le **mouvement est uniforémement accéléré** car la vitesse est une fonction négative, affine de pente décroissante. La vitesse est dans le sens inverse de Ox et l'accélération est également dans le même sens.

## Etude du mouvement 3

Encore une fois, commençons par importer les données et tracer x & y en fonction du temps :

```python
source = open('Mvt3.csv', 'r') 
t, x, y = [], [], [] 
for row in csv.reader(source,delimiter=','):
    ti, xi, yi = map(float,row)
    t.append(ti) 
    x.append(xi) 
    y.append(yi) 

plt.title('x & y en fonction du temps pour le mvt 3')
plt.plot(t, x, 'r', label ="x = f(t)")
plt.plot(t, y, 'b', label ="y = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("position", fontsize=12)
plt.legend()
plt.show()
```

Cette fois-ci nous voyons que le mouvement est plan (car x & y sont non nuls) et la trajectoire :

```python
plt.title('Trajectoire pour le mvt 3')
plt.plot(x, y, 'r', label ="y en fonction de x")
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

Est caractéristique d'un mouvement circulaire.

Pour savoir s'il est uniforme, nous pouvons tracer la vitesse en fonction du temps :

```python
vx, vy, v = [], [], []
vx.append((x[1]-x[0])/(t[1]-t[0]))
vy.append((y[1]-y[0])/(t[1]-t[0]))
v.append((vx[0]**2+vy[0]**2)**0.5)
for i in range(1,len(x)-1): 
    vxi=(x[i+1]-x[i-1])/(t[i+1]-t[i-1])
    vyi=(y[i+1]-y[i-1])/(t[i+1]-t[i-1])
    vx.append(vxi)
    vy.append(vyi)
    v.append((vxi**2+vyi**2)**0.5)
vx.append((x[n-1]-x[n-2])/(t[n-1]-t[n-2]))
vy.append((y[n-1]-y[n-2])/(t[n-1]-t[n-2]))
v.append((vx[n-1]**2+vy[n-1]**2)**0.5)

plt.title('v en fonction du temps pour le mvt 3')
plt.plot(t, v, 'r', label ="v = f(t)")
plt.xlabel("temps (s)", fontsize=12)
plt.ylabel("vitesse", fontsize=12)
plt.legend()
plt.show()
```

Nous voyons, que la vitesse est constante (encore une fois, on ne se soucie pas de ce qui se passe sur les bords en 0 & en n-1).

C'est donc un **mouvement circulaire uniforme**.

Traçons le vecteur vitesse et le vecteur accélération en un point particulier.

```python
ax, ay = [], []
ax.append((vx[1]-vx[0])/(t[1]-t[0]))
ay.append((vy[1]-vy[0])/(t[1]-t[0]))
for i in range(1,n-1):
    axi=(vx[i+1]-vx[i-1])/(t[i+1]-t[i-1])
    ayi=(vy[i+1]-vy[i-1])/(t[i+1]-t[i-1])
    ax.append(axi)
    ay.append(ayi)
ax.append((vx[n-1]-vx[n-2])/(t[n-1]-t[n-2]))
ay.append((vy[n-1]-vy[n-2])/(t[n-1]-t[n-2]))

plt.title('Trajectoire pour le mouvement 3')
plt.plot(x,y, 'r+')
plt.quiver(x[50],y[50],vx[49],vy[49], color='green',angles='xy', units='xy', scale=3)
plt.quiver(x[50],y[50],ax[48],ay[48], color='blue',angles='xy', units='xy', scale=3)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
